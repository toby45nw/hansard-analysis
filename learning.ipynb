## Learning Element Tree

```python
sample = """<speech speakerid="mr-smith" id="s1">
  <p>This is the first paragraph.</p>
  <p>The Right Honourable <member>Mr. Smith</member> rose to speak.</p>
</speech>"""

root = ET.fromstring(sample)

# Example 1
print(root[1].text +root[1][0].text + root[1][0].tail)

# Example 2
def get_full_text(element):
    parts = []
    if element.text:
        parts.append(element.text)
    for child in element:
        parts.append(get_full_text(child))
        if child.tail:
            parts.append(child.tail)
    return ''.join(parts)
print(get_full_text(root[1]))

# Example 3
print("".join(root[1].itertext()))
```
