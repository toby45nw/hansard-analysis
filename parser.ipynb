## Learning Element Tree

```python
sample = """<speech speakerid="mr-smith" id="s1">
  <p>This is the first paragraph.</p>
  <p>The Right Honourable <member>Mr. Smith</member> rose to speak.</p>
</speech>"""

root = ET.fromstring(sample)

# Example 1
print(root[1].text +root[1][0].text + root[1][0].tail)

# Example 2
def get_full_text(element):
    parts = []
    if element.text:
        parts.append(element.text)
    for child in element:
        parts.append(get_full_text(child))
        if child.tail:
            parts.append(child.tail)
    return ''.join(parts)
print(get_full_text(root[1]))

# Example 3
print("".join(root[1].itertext()))
```

```python
file = 'data/debates/debates1919-02-04a.xml'

tree = ET.parse(file)
root = tree.getroot()


for child in root:
    if child.tag == 'speech':
        if 'speakername' in child.attrib:
            print(child.attrib['speakername'], "".join(child.itertext()))

        else:
            print("Unknown speaker", "".join(child.itertext()))
```

## Survey of Files

```python
DEBATES_DIR = Path("data/debates")

files = list(DEBATES_DIR.glob("*.xml"))
print(len(files))
print(files[0])
print(type(files[0]))
print()
print(files[0].name)
print(files[0].stem)
print(files[0].suffix)

print(files[0].stem[7:17])
```

```python
years = []
for f in files:
    years.append(f.stem[7:11])

print(sorted(set(years)))
```

```python
attribs_per_year = {}

for f in files:
    year = f.stem[7:11]

    if year not in attribs_per_year:
        attribs_per_year[year] = set()

    tree = ET.parse(f)
    root = tree.getroot()
    for child in root:
        if child.tag == 'speech':
            attribs_per_year[year].update(child.attrib.keys())
```

```python
for i in attribs_per_year:
    print(i, sorted(attribs_per_year[i]))
```
